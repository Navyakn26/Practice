```python
import pandas as pd
import numpy as np
```

```python
a = pd.read_csv("Sales Transactions-2017.csv")
a
```

```python
b =pd.read_csv("Sales Transactions-2018.csv")
b
```

```python
c = pd.read_csv("Sales Transactions-2019.csv")
c
```

```python
complete_sales_data = complete_sales_data.drop(['Disc','Gross',"Voucher Amount"], axis = 1)
complete_sales_data
```

```python
complete_sales_data.drop_duplicates(inplace=True)
complete_sales_data.shape
```

```python
# replacing empty space by nan
complete_sales_data.replace(" ",np.nan,inplace=True)
complete_sales_data
```

```python
#Droping nan 
complete_sales_data.dropna(subset=["Date","Voucher"],inplace=True)
```
