## Data Preperation 

```python
del claimants_data['CASENUM']
```

```python
claimants_data.head()
```

```python
claimants_data.dropna(inplace=True)
```

```python
claimants_data.isna().sum()
```

```python
claimants_data.dtypes
```

## Model building

```python
x = claimants_data.drop(labels='ATTORNEY',axis=1)
y = claimants_data[['ATTORNEY']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.20,shuffle = True,random_state = None,)
```

```python
x_train.shape,y_train.shape
```

```python
x_train
```

## Model Training

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()
logistic_model.fit(x,y)
```

```python
from sklearn.tree
```

```python
logistic_model.coef_
```

```python
logistic_model.intercept_
```

## Model testing  ||


### Training data

```python
y_pred_train = logistic_model.predict(x_train)
```

### Test Data

```python
y_pred_test = logistic_model.predict(x_test)
```

# Model Evaluation

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
accuracy_score(y_train,y_pred_train)
```

```python
print('classification Report:\n',classification_report(y_train,y_pred_train))
```

```python
print('confusion Matrix : \n',confusion_matrix(y_train,y_pred_train))
```

```python
accuracy_score(y_test,y_pred_test)
```
