### 2. Using Euclideon distance as a Metric for UBCF

```python
from sklearn.metrics import pairwise_distances
```

```python
euclideon_recommendataion = pd.pivot_table(data = user_data,values='rating',index='userId',columns='movie').fillna(0)
euclideon_recommendataion
```

```python
user_data['userId'].unique()
```

```python
euclideon_df = pd.DataFrame(pairwise_distances(X = euclideon_recommendataion,metric='euclidean')).round(2)
euclideon_df.columns = user_data['userId'].unique()
euclideon_df.index   = user_data['userId'].unique()
euclideon_df
```

### 3. Using Cosine distance as a Metric for UBCF

```python
cosine_recommendataion = pd.pivot_table(data = user_data,values='rating',index='userId',columns='movie').fillna(0)
cosine_recommendataion
```

```python
cosine_df = pd.DataFrame(1 - pairwise_distances(X = cosine_recommendataion,metric='cosine')).round(2)
cosine_df.columns = user_data['userId'].unique()
cosine_df.index   = user_data['userId'].unique()
cosine_df
```
